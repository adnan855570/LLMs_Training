```python
from openai import OpenAI

# Set your OpenAI API key
api_key = 'sk-proj-em21xW9CQaFNFmNHgE9OT3BlbkFJFEN0IfatZAmSy4hIfbKb'
client = OpenAI(api_key=api_key)

# Function to classify a tweet using GPT-3.5 or GPT-4
def classify_tweet(tweet):
    # Construct the prompt
    prompt = f"Classify the following Urdu tweet as hate speech or not hate speech:\n\nTweet: \"{tweet}\"\nClassification:"

    # Make a request to the GPT-3.5 or GPT-4 API
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    # Extract and return the classification result
    # Updated to access the content from the 'content' attribute of the message object
    classification = completion.choices[0].message.content.strip()
    return classification

# Function to prompt the user for input and classify the tweet
def prompt_user_for_tweet():
    tweet = input("Please enter an Urdu tweet for classification: ")
    classification = classify_tweet(tweet)
    print(f"Tweet: {tweet}\nClassification: {classification}\n")

# Example usage of predefined tweets
def classify_predefined_tweets():
    tweets = [
        "گانڈو ہم پھاڑتے ہیں ہماری کون پھاڑے گا تم لوگوں کی گانڈ میں اتنا دم نہیں زہنی مریض غلام نواز کے چوتڑ چاٹ یہاں کیا بک بک کرنے آگیا ہے",
        "آج کا موسم بہت اچھا ہے",
        "کسی کے خلاف تشدد کی بات کرنا غلط ہے",
        "مجھے نئی فلم بہت پسند آئی"
    ]

    for tweet in tweets:
        result = classify_tweet(tweet)
        print(f"Tweet: {tweet}\nClassification: {result}\n")

if __name__ == "__main__":
    while True:
        print("1. Classify a user-provided tweet")
        print("2. Classify predefined example tweets")
        print("3. Exit")
        choice = input("Please enter your choice (1/2/3): ") # Take user input for choice

        if choice == "1":
            prompt_user_for_tweet()
        elif choice == "2":
            classify_predefined_tweets()
        elif choice == "3":
            print("Exiting...")
            break
        else:
            print("Invalid choice, please try again.")
```
